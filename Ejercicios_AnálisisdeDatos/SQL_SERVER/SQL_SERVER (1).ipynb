```python
import pyodbc
import pandas as pd
```

```python
conn= pyodbc.connect('Driver={SQL Server};'
                      'Server=DESKTOP-H7FCQF7\SQLEXPRESS;'
                      'Database=Covid;'
                      'Trusted_Connection=yes;'
)
cursor=conn.cursor()
df=pd.read_csv("Ecuador_crisis_covid.csv")
cursor.execute('''
		CREATE TABLE Ecuador_covid_19(
			Fecha nvarchar(10) primary key,
            Incremento_muertos nvarchar(10),
            Muertos nvarchar(10),
            Muertos_millon nvarchar(10),
            Incremento_confirmados nvarchar(10),
            Confirmados nvarchar(10))
               ''')
for row in df.itertuples():
    cursor.execute('''
                INSERT INTO Ecuador_covid_19 (Fecha, Incremento_muertos, Muertos, Muertos_millon, Incremento_confirmados, Confirmados)
                VALUES (?,?,?,?,?,?)
                ''',
                row.Fecha, 
                row.Incremento_muertos,
                row.Muertos,
                row.Muertos_millon, 
                row.Incremento_confirmados,
                row.Confirmados
                   
                )

conn.commit()
cursor.close()
```
